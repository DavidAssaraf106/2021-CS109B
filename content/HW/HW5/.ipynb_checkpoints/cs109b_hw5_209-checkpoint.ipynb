### Overview
During the lecture we discussed about the formulation of the  *Reservoir Computing (RC); an echo-state recurrent neural network*. One of the examples that we discussed was the predictions of the chaotic time-series for the Mackey-Glass nonlinear dynamical system. This example can be found in the seminal paper: http://www.rctn.org/vs265/jaeger04-ESN.pdf. This is the paper that RC was introduced, so we encourage you to read this publication. 

In this homework, you are asked to work on the Rossler dynamical system. It is a very popular chaotic system  used to describe the evolution of chemical reactions. For more information check the Wikipedia:  https://en.wikipedia.org/wiki/R%C3%B6ssler_attractor
For the implementation of the RC you should the python class `pyESN` taken from https://github.com/cknd/pyESN. In this github repository you can also find the Mackey-Glass example. We encourage you to explore this library. Note, `pyESN` can also be found in the current homework directory.

**The task:** Employ the RC network to predict the time evolution of  chaotic time series.  We provide you three time series `x(t), y(t), z(t)` (files: `x.dat`, `y.dat`, `z.dat`), which are solutions of the chaotic Rossler system. 

**In the Question 1**, you are asked to make a short-range forecast. It is a prediction where the network is learning from the past data and predicts the future behavior. In this case, we need a single input. The input signal is a constant function because the sequential steps are considered to be  equally-spaced (difference between time steps), hence, the input should be an array of ones. 

**In the Question 2**  you need to to make a long-time prediction. First, you should use a single constant input (similar to the question 1) where you have to confirm that the predictions are very bad. This is expected because we deal with  very difficult (chaotic) time-series. On the other hand, we saw in the class that  using the concept of the `observers`  we can perform extremely long-range predictions. To include the observers you need to use three different  inputs: a vector of ones (as in the question 1), and the two other known time series `y(t)` and `z(t)` (the observers). In this case the prediction is a response to the past behavior and to the present observers' signals. This kind of predictions are often called **inference**. 

As we discussed in the lecture,  RC is very sensitive to the hyper-parameter. In all the questions you are asked to find  the optimal set of hyper-parameter that gives the best predictions. For your convenience, we are asking you to optimize just two of the hyper-parameter, the `spectrum radius`  and `sparsity term`.  The rest hyper-parameters are given. For this homework you should use a two-dimensional grid to  search the two hyper-parameters, report the mean square error (MSE) for each hyper-parameter set, and keep the model with the lowest MSE.

##### Brief explanation of the parameters:

`n_reservoir`: number of reservoir neurons <br> 
`random_state`: seed for the random generator<br>
`sparsity`: proportion of recurrent weights set to zero <br>
`spectral_radius`: spectral radius of the recurrent weight matrix <br>
`noise`: noise added to each hidden neuron (regularization) <br>

**Helper functions**

We are providing three helper functions.  You can use them or you can write your custom functions. If you define any other helper function (e.g. a function for plotting your results), it should be placed after the next cell. Helper functions are not requested by this homework.


The given helper functions calculate the mean square error `MSE`, the `residuals`, and prepare the data. The `prepareData` splits the data into training and testing sets, and create a training and testing array of ones (account to the time-input).  We do not need evaluation data sets. Note that the `prepareData` **does not** prepare the *observers*, you will need to do it manually in the question 2.

```python
# HELPER FUNCTIONS GO HERE

def myMSE(prediction,target):
    return np.sqrt(np.mean((prediction.flatten() - target.flatten() )**2))

def residuals(prediction,target):
    return (target.flatten() - prediction.flatten())

    
def prepareData(target, train_perc=0.9, plotshow=False):
    datalen =  len(target)        
    trainlen = int(train_perc*datalen)
    testlen  = datalen-trainlen

# Train/Test sets
    trainTarget = target[:trainlen]
    testTarget  = target[trainlen:trainlen+testlen]    
    inputTrain = np.ones(trainlen)
    inputTest  = np.ones(testlen)
        
    if plotshow:
        plt.figure(figsize=(14,3))
        plt.plot(range(0,trainlen), trainTarget,'g',label='Train')
        plt.plot(range(trainlen,trainlen+testlen), testTarget,'-r',label='Test')
        plt.legend(loc=(0.1,1.1),fontsize=18,ncol=2)
        plt.tight_layout()
        
    return trainTarget, testTarget, inputTrain, inputTest

```

**If you define any custom helper function, it should be put in the following cell**

```python
# HELPER FUNCTIONS GO HERE
### your code here
def plotResults(trainlen, testlen, data, yfit, yhat, resTrain, resTest):
    plt.figure(figsize=(14,6))
    plt.subplot(2,1,1)
    plt.plot(range(0,trainlen+testlen), data,'k',  linewidth=2, label="data")
    plt.plot(range(0,trainlen),yfit,'--g',  linewidth=2, alpha=0.9, label="train")
    plt.plot(range(trainlen,trainlen+testlen), yhat,'--r', linewidth=2,  alpha=1, label="test")
    lo,hi = plt.ylim()
    plt.plot([trainlen,trainlen],[lo+np.spacing(1),hi-np.spacing(1)],'--b',alpha=0.8, linewidth=4)
    plt.tight_layout()
    plt.ylabel('x')
    plt.legend(loc=(0.1,1.1),fontsize=18,ncol=3)

    
    #
    plt.subplot(2,1,2)
    plt.plot(range(0,trainlen),resTrain,'--g',  linewidth=2, alpha=0.9, label="train")
    plt.plot(range(trainlen,trainlen+testlen), resTest,'--r', linewidth=2,  alpha=1, label="test")
    lo,hi = plt.ylim()
    plt.plot([trainlen,trainlen],[lo+np.spacing(1),hi-np.spacing(1)],'--b',alpha=0.8, linewidth=4)
    plt.tight_layout()
    plt.ylabel('Residuals')
```

#### 1.1 Manual hyper-parameter search and prediction

Set the target time-series and give the name `target`

```python
### your code here
target=x
```

**Prepare your data:** The target time-series should be splitted  into training and testing sets. Plot the time-series and use different colors to indicate the training and testing sets.  You might want to use the given helper function `prepareData()` or you can do it by yourself.

```python
### your code here
trainTarget, testTarget, inputTrain, inputTest = prepareData(target, train_perc=0.95, plotshow=True)
```

Make a quick prediction until find a testing MSE < 0.25: Try around `spectral_radius = [1.2, 2.6]` and   `sparsity = [0.16,  0.24]`. 


```python
### your code here
n_outputs = 1
n_reservoir = 1000
noise = 0.0001
random_state=42
```

```python
spectral_radius = [1.2, 2.6]
sparsity = [0.16, 0.24]

esn = ESN(n_inputs = 1,
      n_outputs = n_outputs,
      n_reservoir = n_reservoir,
      noise = noise,
      spectral_radius = 1.2,
      sparsity=0.16,          
      random_state=random_state)
yfit = esn.fit(inputTrain, trainTarget)
yhat = esn.predict(inputTest)
mse = myMSE(yhat, testTarget)
resTrain = residuals(yfit, trainTarget)
resTest = residuals(yhat, testTarget)
```

```python
print(f"For the set of parameters: spectral_radius = 1.2 & sparsity = 0.16, the MSE value on the test set it: {mse:.4f}")
```

```python
trainlen = int(len(x)*0.95)
testlen = int(len(x)*0.05)
plotResults(trainlen, testlen, x, yfit, yhat, resTrain, resTest)
```

#### 1.2 Hyper-parameters optimization
Make a grid search on the hyper-parameters `spectra-radius` and `sparsity`. Visualize the result by ploting the testing MSE in a 2D color plot.

**Tips:** Search in the range: `spectral_radius = [1.2, 2.6]` and   `sparsity = [0.16,  0.24]`. The RC is more sensitive on the `sparsity`. For a dense grid you can consider the steps 0.05 and 0.01 for the `spectra-radius` and `sparsity`, respectively.

```python
### your code here
spectral_radius_list = np.arange(1.2,2.6,0.2)
sparsity_list = np.arange(0.16,0.24,0.01)

mse_lst = []
yfit_lst = []
yhat_lst = []
resTrain_lst = []
resTest_lst = []
sp_rad_spar_lst = []

for sp_rad in spectral_radius_list:
    for spar in sparsity_list:
        esn = ESN(n_inputs = 1,
                  n_outputs = n_outputs,
                  n_reservoir = n_reservoir,
                  noise = noise,
                  spectral_radius = sp_rad,
                  sparsity=spar,          
                  random_state=random_state)
        yfit = esn.fit(inputTrain, trainTarget)
        yfit_lst.append(yfit)
        yhat = esn.predict(inputTest)
        yhat_lst.append(yhat)
        sp_rad_spar_lst.append([sp_rad,spar])
        mse = myMSE(yhat, testTarget)
        mse_lst.append(mse)
        resTrain = residuals(yfit, trainTarget)
        resTrain_lst.append(resTrain)
        resTest = residuals(yhat, testTarget)
        resTest_lst.append(resTest)
```

**plot the results**

```python
### your code here
fig, ax = plt.subplots(figsize = (12,12))
ax.set_xlabel('Sparsity', fontsize = 12)
ax.set_ylabel('Spectral Radius', fontsize = 12)
ax.set_title('Figure: Heatmap for the Gridsearch of Hyperparameters: Spectral Radius & Sparsity', fontsize = 16)
im = ax.imshow(np.array(mse_lst).reshape(8, 7))
cbar = ax.figure.colorbar(im, ax=ax)

plt.xticks(range(len(sparsity_list)), [round(float(spa),2) for spa in sparsity_list])
plt.yticks(range(len(spectral_radius_list)), [round(float(spa),2) for spa in spectral_radius_list])
plt.tight_layout()
plt.show()
```

#### Optimal prediction

Plot the training and the prediction along with the ground truth data. Also, show the residual between the ground truth and your prediction.

```python
### your code here
min_idx = np.argmin(mse_lst)
yfit =  yfit_lst[min_idx]
yhat = yhat_lst[min_idx]
resTrain = resTrain_lst[min_idx]
resTest = resTest_lst[min_idx]

plotResults(trainlen, testlen, x, yfit, yhat, resTrain, resTest)
```

```python
print(f"For the set of parameters: spectral_radius: {np.round(sp_rad_spar_lst[min_idx][0], 2)} & sparsity: {np.round(sp_rad_spar_lst[min_idx][1], 2)}, the MSE value on the test set it: {mse_lst[min_idx]:.4f}")
```
